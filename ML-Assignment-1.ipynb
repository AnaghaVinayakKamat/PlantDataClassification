```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

x_train_data = plant_train[['sepal length in cm', 'sepal width in cm', 'petal length in cm', 'petal width in cm']]
y_train_data = plant_train[['target/label']]

x_test_data = plant_test[['sepal length in cm', 'sepal width in cm', 'petal length in cm', 'petal width in cm']]
y_test_data = plant_test[['target/label']]
```

# Model 1 SVM

```python
from sklearn.svm import SVC

classifier = SVC()
classifier.fit(x_train_data, y_train_data)

y_pred = classifier.predict(x_test_data)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test_data, y_pred))
```

```python
cm = confusion_matrix(y_test_data, y_pred)
sns.heatmap(cm, annot=True)
```

```python
print('accuracy is',accuracy_score(y_pred,y_test_data))
```
